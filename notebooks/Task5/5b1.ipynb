```python
import json, os, re
import pandas as pd
```

```python
def get_title(filepath):
    f = open(filepath)
    data = json.load(f)
    name = data['Message:From-Name']
    name_new = re.sub('[^A-Za-z0-9]+', ' ', name)
    name_new = name_new.upper()
    name_new = name_new.replace('FROM','')
    ls = name_new.split()
    return ls[0]
```

```python
title = {}
cnt = len(os.listdir('data/separated by email'))
for i in range(cnt):
    filepath = 'data/separated by email/' + str(i) + '.json'
    try:
        title[i] = get_title(filepath)
    except:
        title[i] = 'N/A'
freq = {}
for num, tit in title.items():
    
    if tit in freq.keys():
        freq[tit] += 1
    else:
        freq[tit] = 1
df = pd.DataFrame(freq.items(), columns = ['title','count'])
df = df.sort_values(['count'], ascending = False)
```

```python
freq_new = freq
list = ['MRS','MS','MISS','MR','PRINCE','PRINCESS']
for i in list:
    for titl,cnt in freq_new.items():
        if len(i)<len(titl) and i in titl and titl != 'MRS':
            print([i,titl])
            if freq_new[titl] >= 1:
                freq_new[i] = freq_new[i] + 1
                freq_new[titl] = freq_new[titl] - 1
df2 = pd.DataFrame(freq_new.items(), columns = ['title','count'])
df2 = df2.sort_values(['count'], ascending = False)
```
