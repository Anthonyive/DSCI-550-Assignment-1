```python
import json, os, re
import pandas as pd
```

```python
def get_title(filepath):
    f = open(filepath)
    data = json.load(f)
    name = data['Message:From-Name']
    name_new = re.sub('[^A-Za-z0-9]+', ' ', name)
    name_new = name_new.upper()
    name_new = name_new.replace('FROM','')
    ls = name_new.split()
    return ls[0]
```

```python
title = {}
cnt = len(os.listdir('data/separated by email'))
for i in range(cnt):
    filepath = 'data/separated by email/' + str(i) + '.json'
    try:
        title[i] = get_title(filepath)
    except:
        title[i] = 'N/A'
freq = {}
for num, tit in title.items():
    
    if tit in freq.keys():
        freq[tit] += 1
    else:
        freq[tit] = 1
df = pd.DataFrame(freq.items(), columns = ['title','count'])
df = df.sort_values(['count'], ascending = False)
```

```python
title2 = title.copy()
list = ['MRS','MS','MISS','MR','PRINCE','PRINCESS','DR','LAWYER','PROFESSOR','ENGINEER']
for i in list:
    for num, name in title2.items():
        if name.startswith(i,0,len(i)):
            title2[num] = i

for num,name in title2.items():
    if name in ['MRS','MR','MISS','MS']:
        title2[num] = 1
    elif name in ['PRINCE','PRINCESS']:
        title2[num] = 2
    elif name in ['DR','LAWYER','PROFESSOR','ENGINEER']:
        title2[num] = 3
    elif name == 'N/A':
        title2[num] = 0
    else:
        title2[num] = 4
df_t = pd.DataFrame(title2.items(),columns = ['filenumber','index'])
```
