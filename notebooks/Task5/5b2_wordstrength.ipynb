```python
import json
import os
from nltk import WordPunctTokenizer
wpt = WordPunctTokenizer()

datafiles = 'data'

def readfile(filepath):
    doc=''
    if filepath.endswith('.txt'):
        with open(filepath,'rb') as f:
            lines=[str(l)[2:-1] for l in f.readlines()]
        doc = ' '.join(lines)
    elif filepath.endswith('.json'):
        tmp = json.load(open(filepath,'r',encoding='utf8'))
        if type(tmp)==type({}):
            doc = tmp.get('X-TIKA:content','')+' '+ tmp.get('dc:title','')
        else:
            lines = [l['X-TIKA:content'] for l in tmp if 'X-TIKA:content' in l.keys()]
            titles = [l['dc:title'] for l in tmp if 'dc:title' in l.keys()]
            doc = ' '.join(lines) +' '+ ' '.join(titles)
    return doc

def count(email):
    email = email.strip()
    if email=='':
        return 0,0
    email = email.lower()
    tokens = wpt.tokenize(email)
    nurgent = tokens.count('urgent')
    nnow = tokens.count('now')
    return nurgent,nnow
            
data = {}
total_urgent_frequency,total_now_frequency = 0,0
for file in os.listdir(datafiles):
    filedir = os.path.join(datafiles,file)
    if os.path.isfile(filedir):
        pass #if we don't need to count files outside of "separated by email", then pass this if statement
#         sample = readfile(filedir)
#         counts = {}
#         counts['urgent_frequency'],counts['now_frequency'] = count(sample)
#         total_urgent_frequency+=counts['urgent_frequency']
#         total_now_frequency+=counts['now_frequency']
#         data[file] = counts
#         print(file,counts)
    else:
        for subfile in os.listdir(filedir):
            subfiledir = os.path.join(filedir,subfile)
            sample = readfile(subfiledir)
            counts = {}
            counts['urgent_frequency'],counts['now_frequency'] = count(sample)
            total_urgent_frequency+=counts['urgent_frequency']
            total_now_frequency+=counts['now_frequency']
            data[subfile[:-5]] = counts
            print(subfile[:-5],counts)
# counts = {}
# counts['urgent_frequency'],counts['now_frequency'] = total_urgent_frequency,total_now_frequency          
# data['total'] = counts
# print('total',counts)
with open('results.json', 'w') as result_file:
    json.dump(data, result_file)
```
