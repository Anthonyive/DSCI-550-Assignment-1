```python
import os, json, re
from spellchecker import SpellChecker
```

```python
def spell(filepath):
    cnt = 0
    f = open(filepath)
    data = json.load(f)
    content = data['X-TIKA:content']
    content_new = re.sub('[^A-Za-z]+', ' ', content)
    ct = content_new.split()
    spell = SpellChecker()
    for i in ct:
        if i not in spell:
            cnt += 1
    return cnt
```

```python
spell('data/separated by email/1.json')
```

```python
cnt_file = len(os.listdir('data/separated by email'))
dict = {}

for i in range(1,cnt_file):
    filepath = 'data/separated by email/' + str(i) + '.json'
    try:
        num = spell(filepath)
    except:
        num = 'N/A'
    dict[i] = num
```
