```python
import json

j = json.load(open('../../data/fradulent_emails_t.json'))
df = pd.json_normalize(j, max_level=0)
```

```python
list(df.columns)
```

```python
def case_insensitive_unique_list(data):
    seen, result = set(), []
    for item in data:
        if item.lower() not in seen:
            seen.add(item.lower())
            result.append(item)
    return result
```

```python
len(case_insensitive_unique_list(df.columns))
```
