```python
import json
import re
import requests
import pandas as pd
```

```python
df = pd.read_csv(f'../../data/s_p_500_historical_data.csv')
```

```python
def edit_date_format(original):
    '''original -> str, original date format xxxx-xx-xx
       output ret -> str, new format mm/dd/yy'''
    spl = original.split('-')
    mm, dd, yy = spl[1], spl[2], spl[0][2:]
    ret = f'{mm}/{dd}/{yy}'
    
    return ret
```

```python
# edit date format
df['Date'] = df['Date'].apply(edit_date_format)
# remove unrelated columns
df = df.drop(columns=['High', 'Low', 'Adj Close'])
# rename columns to lowercase
df = df.rename(columns={'Date': 'date', 'Open': 'open', 'Close': 'close', 'Volume': 'volume'})
# convert to json
d_dict = json.loads(df.T.to_json())
```
