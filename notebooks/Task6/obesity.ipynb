```python
def get_info(index):
    # COllect age and date information from existed data.
    info = json.load(open('dataset.json'))
    # info.items()

    date = info['assignment-1']['task-5-additional-features'][index]['date']
    temp = date.split('/')
    yr = temp[2]
    age = info['assignment-1']['task-5-additional-features'][index]['attacker estimated age']['age prediction']
    return yr,age
```

```python
def survey_years(yr):
    yr = int(yr)
    if yr > 90:
        if yr %2 == 0:
            year = f'19{yr-1}-19{yr}'
        else:
            year = f'19{yr}-19{yr+1}'
    elif yr == 9 or yr == 10:
        year = '2009-2010'
    elif yr >= 11:
        if yr %2 == 0:
            year = f'20{yr-1}-20{yr}'
        else:
            year = f'20{yr}-20{yr+1}'
    elif yr%2 == 0:
        year = f'200{yr-1}-200{yr}'
    else:
        year = f'200{yr}-200{yr+1}'
    return year
year = survey_years(yr)
year
# Make year into its related format. For example, change 02 into 2001-2002
```

```python
def age_group(age):
    if 20 <= age < 39:
        group = '20-39'
    elif 39 <= age < 60:
        group = '40-59'
    elif age >= 60:
        group = '60 and over'
    else:
        group = 'N/A'
    return group
# Make age into its suitable age group.
group = age_group(float(age[0]))
```

```python
data_temp = df.loc[(df['Survey Years']==year) & (df['Age Group'] == group) & (df['Sex'] == 'All')]
# FIlter the pandas dataframe to get the info I need
```

```python
obesity = data_temp['Percent'][data_temp['Measure']=='Obesity'].values[0]
```

```python
hyper = data_temp['Percent'][data_temp['Measure']=='Hypertension'].values[0]
```

```python
HTC = data_temp['Percent'][data_temp['Measure']=='High Total Cholesterol'].values[0]
```

```python
output = {}
```

```python
for i in tqdm(range(4029,4030)):
    tem_dict = {}
    try:
        yr,age = get_info(i)
        age = float(age[0])
        year = survey_years(yr)
        group = age_group(age)
        data_temp = df.loc[(df['Survey Years']==year) & (df['Age Group'] == group) & (df['Sex'] == 'All')]
        obesity = data_temp['Percent'][data_temp['Measure']=='Obesity'].values[0]
        hyper = data_temp['Percent'][data_temp['Measure']=='Hypertension'].values[0]
        HTC = data_temp['Percent'][data_temp['Measure']=='High Total Cholesterol'].values[0]
        tem_dict['Obesity Rate'] = obesity
        tem_dict['Hypertension Rate'] = hyper
        tem_dict['High Total Cholesterol Rate'] = HTC
        output[i] = tem_dict
    except:
        # In case the year and age are not available for some specific emails.
        tem_dict['Obesity Rate'] = 'N/A'
        tem_dict['Hypertension Rate'] = 'N/A'
        tem_dict['High Total Cholesterol Rate'] = 'N/A'
        output[i] = tem_dict
#
```
